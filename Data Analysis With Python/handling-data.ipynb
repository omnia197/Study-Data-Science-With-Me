# Data Cleaning

```python
data[data['County'].isnull() == True]
#the 4 column has null values at County, city , Vehicle Location ,Electric Utility ,2020 Census Tract
# sooo , we can drop these rows
```

```python
data.dropna(subset = 'County' , inplace = True)
#rename the column 
data.rename(columns = {'County' : "country"} , inplace = True)
```
